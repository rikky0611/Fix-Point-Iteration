## e^x-sinx-3x+1=0の根を不動点反復で数値的に求める。

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pylab
```

### 不動点反復
x = g(x)の形にする。不動点反復形は３つ考えられる。


```python
def function_1(x):
    return np.arcsin(np.exp(x)-3*x+1)

def function_2(x):
    return np.log(np.sin(x)+3*x-1)

def function_3(x):
    return (np.exp(x)-np.sin(x)+1)/3

pylab.figure(figsize=(10, 10),dpi=20)
xmin = 0.5
xmax = 1.6
x = np.arange(xmin,xmax,0.001)
y_1 = function_1(x)
y_2 = function_2(x)
y_3 = function_3(x)

plt.subplot(3, 1, 1)
plt.ylim(0.5,1.6)
plt.plot(x,y_1)
plt.plot(x,x)
plt.subplot(3, 1, 2)
plt.ylim(0,2)
plt.plot(x,y_2)
plt.plot(x,x)
plt.subplot(3, 1, 3)
plt.ylim(0,2)
plt.plot(x,y_3)
plt.plot(x,x)
```

### function_1について
これは微分係数(後述)が大きいために収束までに時間がかかる。(120回くらいの反復を必要とした)

```python
def fixed_point_iteration(function, x_0, epsilon=1e-5):
    i = 0
    x = x_0
    x_arr.append(x)
    y_arr.append(0)
    while x - function(x) > epsilon or function(x)-x > epsilon:xxxxxx
        print("episode{0}:x={1}:y={2}".format(i,x,function(x)))
        x_arr.append(x)
        y_arr.append(function(x))
        x = function(x)
        x_arr.append(x)
        y_arr.append(x)
        i += 1
        
x_arr = []
y_arr = []
fixed_point_iteration(function_1,1.5)
plt.ylim(0.5,1.6)
plt.plot(x,y_1)
plt.plot(x,x)
plt.plot(x_arr,y_arr)
```

### 微分係数の可視化
x=g(x)としたときのg'(x)の値を可視化する。  
また収束可能性を調べるために、y=1, y=-1も波線にてプロットする。

```python
def deriv_function_1(x):
    denom = (1-(np.exp(x)-3*x+1)**2)**0.5
    return (np.exp(x)-3)/denom

def deriv_function_2(x):
    denom = np.sin(x)+3*x-1
    return (np.cos(x)+3)/denom

def deriv_function_3(x):
    return (np.exp(x)-np.cos(x))/3

xmin = -3
xmax = 3
x = np.arange(xmin,xmax,0.001)
y_1 = deriv_function_1(x)
y_2 = deriv_function_2(x)
y_3 = deriv_function_3(x)

plt.hlines([-1, 1], xmin, xmax, linestyles="dashed")
plt.ylim(-10,10)
plt.plot(x,y_1)
plt.plot(x,y_2)
plt.plot(x,y_3)
```
